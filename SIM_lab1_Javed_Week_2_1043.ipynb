**1.Create Your Personal Matrix**

● d1 = last digit of your student ID

● d2 = second last digit of your student ID

Create the matrix:

**2.Analyze Matrix A**

Using NumPy, compute:

● Shape of A

● Determinant

● Rank

● Eigenvalues

● Inverse (only if determinant ≠ 0)

```python
A = np.array([[5, 5],[6, 6]])
size_A = A.shape
det_A = np.linalg.det(A)
rank_A = np.linalg.matrix_rank(A)
eigenvalues, eigenvector = np.linalg.eig(A)

print("Shape of A Matrix: \n",size_A,"\n")
print("Determinant: \n",det_A, "\n")
print("Rank of Matrix: \n",rank_A,"\n")
print("Eigenvalues of A: \n",eigenvalues,"\n")
print("Eigenvector of A: \n",eigenvector,"\n")
```

```python
B = np.array([[6, 5],[6, 6]])

size_B = B.shape
det_B = np.linalg.det(B)
rank_B = np.linalg.matrix_rank(B)
eigenvalues, eigenvector = np.linalg.eig(B)
if np.linalg.det(B) != 0:
  inv_B = np.linalg.inv(B)
else:
  inv_B = "Matrix is singular,cannot compute "

print("Shape of B Matrix: \n",size_B,"\n")
print("Determinant: \n",det_B, "\n")
print("Rank of Matrix: \n",rank_B,"\n")
print("Eigenvalues of B: \n",eigenvalues,"\n")
print("Eigenvector of B: \n",eigenvector,"\n")
print("Inverse Matrix: \n",inv_B,"\n")
```
