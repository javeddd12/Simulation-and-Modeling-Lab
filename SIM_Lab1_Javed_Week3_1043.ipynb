```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

**Assignment 1**
1. Create two 3Ã—3 matrices with random integers in Python.
2. Perform addition, subtraction, and multiplication operations on these matrices.
3. Compute the determinant, inverse, and rank of one of the matrices.

```python
A = np.random.randint(0, 10, size=(3, 3))
B = np.random.randint(0, 10, size=(3, 3))

print("Matrix A:\n", A)
print("Matrix B:\n", B)

matrix_sum = A + B  # Matrix addition
matrix_diff = A - B  # Matrix subtraction
matrix_prod = np.dot(A, B)  # Matrix multiplication (dot product)

det_A = np.linalg.det(A)
if np.linalg.det(A) != 0:
    inv_A = np.linalg.inv(A)
else:
    inv_A = "Matrix is singular, cannot compute inverse"
rank_A = np.linalg.matrix_rank(A)

print("Addition of matrix: \n",matrix_sum)
print("Subtraction of matrix: \n",matrix_diff)
print("Multiplication of matrix: \n",matrix_prod)

print("Determinant of A:", det_A)
print("Inverse of A:\n", inv_A)
print("Rank of A:", rank_A)
```

1.Generate and Plot Two Random Vectors:

a. Create two vectors, each containing 15 random floats of your choice.

b. Plot these vectors on the same graph using matplotlib to compare how
they look.

c. Label the axes appropriately and give the plot a title.

```python
# Create two vectors, each containing 15 random floats
vector1 = np.random.uniform(0, 10, 15)
vector2 = np.random.uniform(0, 10, 15)

print("Vector 1:\n", vector1)
print("\nVector 2:\n", vector2)

# Plot the vectors on the same graph
plt.figure()
plt.plot(vector1, marker='o', linestyle='-', label="Vector 1")
plt.plot(vector2, marker='s', linestyle='--', label="Vector 2")

# Label the axes
plt.xlabel("Index")
plt.ylabel("Value")

# Add a title
plt.title("Comparison of Two Random Float Vectors")

# Add legend
plt.legend()
plt.grid(True)

plt.show()
```

2.Generate and Visualize a Random Matrix:

a. Create a 4x4 matrix with random values of your choice.

b. Display the matrix using a heatmap with appropriate color mapping (e.g.,
seaborn or matplotlib).

c. Add labels to the heatmap to indicate row and column numbers.

```python
#Create a 4x4 matrix with random values
matrix = np.random.rand(4, 4)
print("4x4 Random Matrix:\n", matrix)

# Step 2: Plot the matrix as a heatmap
sns.heatmap(matrix, annot=True, cmap='viridis',linewidths=2, linecolor='black',
            xticklabels=['Col 1', 'Col 2', 'Col 3', 'Col 4'],
            yticklabels=['Row 1', 'Row 2', 'Row 3', 'Row 4'])

#Add a title
plt.title('Heatmap of 4x4 Matrix')

# Step 4: Show the plot
plt.show()
```

3.Matrix Operations and Visualization:

a. Create two 4x4 matrices with random values of your choice.

b. Perform the following operations:

i. Matrix Addition

ii. Matrix Subtraction

iii. Matrix Multiplication (ensure matrices are compatible for
multiplication)

c. For each of these operations, visualize the result using bar plots. Each bar
should represent one element of the matrix.

```python
# a. Create two 4x4 matrices with random integer values between 0 and 10
A = np.random.randint(0, 10, size=(4, 4))
B = np.random.randint(0, 10, size=(4, 4))

print("Matrix A:\n", A, "\n")
print("Matrix B:\n", B, "\n")

# b. Perform matrix operations
matrix_sum = A + B          # i. Matrix Addition
matrix_diff = A - B         # ii. Matrix Subtraction
matrix_prod = np.dot(A, B)  # iii. Matrix Multiplication

# c. Visualize the results using bar plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))  # 1 row, 3 columns for plots

# Flatten matrices to make 1D array for plotting
indices = range(1, 17)  # 4x4 = 16 elements

# Matrix Addition
axes[0].bar(indices, matrix_sum.flatten(), color='blue')
axes[0].set_title('Matrix Addition')
axes[0].set_xlabel('Element Index')
axes[0].set_ylabel('Value')

# Matrix Subtraction
axes[1].bar(indices, matrix_diff.flatten(), color='green')
axes[1].set_title('Matrix Subtraction')
axes[1].set_xlabel('Element Index')
axes[1].set_ylabel('Value')

# Matrix Multiplication
axes[2].bar(indices, matrix_prod.flatten(), color='red')
axes[2].set_title('Matrix Multiplication')
axes[2].set_xlabel('Element Index')
axes[2].set_ylabel('Value')

plt.tight_layout()
plt.show()
```
